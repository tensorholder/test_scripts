```python
def loged(str):
    log = open(absolutdir+'log','a+');
    log.write(str+'\n')
    log.close()

def read_log():
    log = open(absolutdir+'log','r+');
    return [line.rstrip('\n') for line in log.readlines()]

def exception_log(str):
    exp = open(absolutdir+'exception','a+');
    exp.write(str+'\n')
    exp.close()

chk_box = ["chkCompat%s" %i for i in [1,2,3,4]]
email = "input#Text1"
password = "input#Password1"
ok = "input#Submit1"
login = "goldofcoin@yandex.ru"
passwd = "W3stbr00k1331"

def log_in():
    driver = webdriver.Firefox()
    driver.get(surl)
    time.sleep(2)
    for chkName in chk_box:
        driver.find_element_by_name(chkName).click()
    driver.find_element_by_name("Submit1").click()
    time.sleep(2)
    driver.find_element_by_css_selector(email).send_keys(login)
    driver.find_element_by_css_selector(password).send_keys(passwd)
    driver.find_element_by_css_selector(ok).click()
    time.sleep(10)
    for chkName in chk_box:
        driver.find_element_by_name(chkName).click()
    driver.find_element_by_name("Submit1").click()
    time.sleep(2)
    driver.find_element_by_name("cmOk").click()
    time.sleep(2)
    return driver

def check_names(et,tmp):
    xs = []
    for t in tmp:
        if t not in et:
            xs.append(t)
    return xs
            

def move_page(num,driver):
      for i in range(num):
            WebDriverWait(driver, 30).until(
                EC.presence_of_element_located(
                (By.XPATH,"//input[@id='Submit3'and @name='cmdNextPage']"))).click()
      return driver

def number_objs():
    sum_ = 0
    for path in os.listdir(absolutdir):
        sum_+= os.path.isdir(absolutdir+'/'+path) and (bool)(os.listdir(absolutdir+'/'+path))
    return sum_

```
