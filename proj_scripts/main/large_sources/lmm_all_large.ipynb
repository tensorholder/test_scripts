```python
import os
import pandas as pd
import numpy as np
import subprocess
import re
import pickle
from make import make_dir,recover_make,all_,comp_,clang_flags,make_proc,make_routine
from make import opt_bc,opt_routine,targ_routine,phony_routine,clean_routine,select_groups
opt_ = ["-O%s" %i for i in [0,1,2,3,'s','z']]

clang_ =["-O%s" %i for i in [0,1,2,3,'fast','s','z','g','4']]

frame = pd.DataFrame(columns=opt_,index=clang_)
absdir = "./../../../../Unix_Source/"
initdir = os.getcwd()+'/'
largedir = absdir+'unix_cmp'
```

```python

reg_name = re.compile(r'.+\s*\:\s*(.+\.[olbcpxCP]{1,3}\s{0,1})+\n')


all_divide = r'all\s*\:\s*'
w_out_space_divide = r'(?<!\\)\s+'
name_divide = r'.+\s*\:\s+.*\.[cpx]{1,3}.*'

mopt_routine = lambda x,y : "{1}:{0}\n\t\topt -Oz {0} -o {1}\n".format(x,y)
```

```python
os.chdir(initdir)
res__ = pickle.load(open('all_single.result','rb'))
```

```python
def cmp_cond(line,xs):
    res = None
    if comp_.match(line):
        cmp = re.split(w_out_space_divide,line)
        cmp = list(filter(lambda x:x!='',cmp))

        if re.match(r'.*\.[cpCP]{3}\s.+',line):
            cmp[0] = 'clang++'
        else:
            cmp[0] = 'clang'
        count = 0
#         print(cmp)
        for ind,el in enumerate(cmp):
            if el[-3:] =='\.cpp':
                count+=1
        if count>2:
            print(line)
        
        tmp = cmp[:2] + cmp[cmp.index('-o'):]

        pos = cmp[cmp.index('-o')+1][:-3]
        if '/' in pos:
#             print(pos)
            xs['target'][pos.split('/')[-1]]['abs_name'] = pos
            pos = pos.split('/')[-1]
            
        xs['target'][pos]['cmp'] = tmp            
#         cmp.insert(2,clang_flags)
#         cmp[cmp.index('-o')+1]+='.ll'
        
        xs['text'].append('\t\t'+' '.join(cmp)+'\n')
#         print('\t\t'+' '.join(cmp)+'\n')
        res = xs
    return res

def all_cond(line,xs):
    res= None
    if all_.match(line):
#         print(xs)
        al = re.split(w_out_space_divide,re.split(r'all\s*\:\s*',line)[1])
        for el in al:
            if el[-3:] != '.bc' and el[-3:]!= '.ll' and el!='\n' and el!='':
                xs['target'].setdefault(el,{'cmp':None,'opt':None,'abs_name':None,'r_name':None})
        xs['text'].append(line)    
        res = xs
    return res

def name_cond(line,xs):
    res = None
    if reg_name.match(line):
#          print(xs)
        name_,adr = re.split(r'\s*:\s*',line)
        xs['target'][name_[:-3]]['abs_name'] = name_[:-3]
#         print(name_)
#         tmp = '{}: {}'.format(name_+'.ll',adr)
        xs['text'].append(line)
        res = xs
    return res
    

                    

def log_file(lines,name=initdir+'multiple.large'):
    with open(name,'a+') as file:
        for line in lines:
            if line[-1]!='\n':
                file.write(line+'\n')
            else:
                file.write(line)
```

```python
for ult in multiple:
    _all = ['all:']
    flag = True
    for targ in multiple[ult]['target']:
        _all+=[targ,multiple[ult]['target'][targ]['abs_name']+'.ll',multiple[ult]['target'][targ]['abs_name']+'.bc']
    _all = ' '.join(_all) +'\n'

    multiple[ult]['text'][0] = _all
    multiple[ult]['text'].insert(0,phony_routine)

    for targ in multiple[ult]['target']:
        multiple[ult]['text'].append(mopt_routine(multiple[ult]['target'][targ]['abs_name']+'.ll'
                                                ,(multiple[ult]['target'][targ]['abs_name']+'.bc')))
        multiple[ult]['opts'].append(multiple[ult]['text'].__len__()-1)
        tmp = multiple[ult]['target'][targ]['cmp']
        if tmp:
            tmp = '{} {} -o {} {}'.format(tmp[0],multiple[ult]['target'][targ]['abs_name']+'.bc',
                                              multiple[ult]['target'][targ]['abs_name'],
                                              ' '.join(tmp[tmp.index('-o')+2:]))
#             print(tmp)
            tmp = '{}: {}\n\t\t{}\n'.format(multiple[ult]['target'][targ]['abs_name'],
                                    multiple[ult]['target'][targ]['abs_name']+'.bc',
                                    tmp)
#             print(tmp)
            multiple[ult]['text'].append(tmp)
    multiple[ult]['text'].append(clean_routine(re.split(all_divide,_all)[1]))
```

```python
err = set()
res_opt = {}
for ult in multiple:
    os.chdir(absdir+'unix_cmp')
    dir_ = make_dir("{}/{}/".format(ult,'source'))[0]
    os.chdir(dir_)
    res_opt.setdefault(ult,{'-O0':{},'-O1':{},'-O2':{},'-O3':{},'-Os':{},'-Oz':{}})
    for option in opt_:
        for op in multiple[ult]['opts']:  
            _opt = (multiple[ult]['text'][op])
            tmp = _opt.split('\n')[1]
            tmp = re.split(w_out_space_divide,tmp)
            tmp[2] = option
            multiple[ult]['text'][op] = '\n'.join([_opt.split('\n')[0],'\t\t'+' '.join(tmp)+'\n'])
        with open('Makefile','wt') as make_file:
            for line in multiple[ult]['text']:
                make_file.write(line)    
        make_proc(['make'],log_file,ult)
        res = -1
        for targ in multiple[ult]['target']:
            try:
                res =subprocess.check_output("wc -c {}".format(multiple[ult]['target'][targ]['abs_name']), stderr=subprocess.STDOUT,
                                                       shell=True)
                res_opt[ult][option][multiple[ult]['target'][targ]['abs_name']] = res
                print(res)
            except Exception:
                print(targ)
                err.add(ult)        
        make_proc(['make','clean'],log_file,ult)
```
